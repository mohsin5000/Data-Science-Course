## Assignment # 1

#### Add Column in dataset 
#### col of cast num_of_cast

```python
import numpy as np
data['cast'] = data['cast'].replace(np.nan, "None")
data.head(2)
```

```python
def num_of_cast(cast):
    return len(cast.split(", "))
```

```python
data["Number Of Cast"] = data["cast"].apply(num_of_cast)
```
